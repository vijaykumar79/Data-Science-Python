# Univariate distribution plot
Plot the data one variable around the categorical variable

## Strip plot

```python
plt.subplot(2,1,1)
sns.stripplot(x='day', y='tip', data=tips) # x-> numeric values y-> categorical value
plt.subplot(2,1,2)
sns.stripplot(x='day', y='tip', data=tips, jitter=True, size=3) #jitter-> remove the overlay data, size-> small the point size
plt.show()
```

## Swarn Plot
Jitter not helping lot to overcome overlap data even with smaller size. Swarn plot is usefull to spread the size of point

```python
plt.subplot(2,1,1)
sns.swarmplot(x='day', y='tip', data=tips)
plt.subplot(2,1,2)
sns.swarmplot(x='day', y='tip', data=tips, hue='sex')
plt.show()
```

## Violin plot and Box Plot
Strip and Swarn plot is usefull for small set of data but for large data it will overlap and then we need violin or boxplot for better descriptive stastic visulaization

```python
sns.violinplot(x='day', y='tip', data=tips)
```

```python
sns.violinplot(x='day', y='tip', data=tips, inner=None) # inner-> remove the inner boxplot
```

```python
sns.violinplot(x='day', y='tip', data=tips, inner=None, color='lightgrey')
```

```python
sns.boxplot(x='day', y='tip', data=tips)
```

```python
# overlap stripplot with violin
sns.violinplot(x='day', y='tip', data=tips, inner=None, color='lightgrey')
sns.stripplot(x='day', y='tip', data=tips, jitter=True, size=2)
```

# Multivariate distribution plot

### Joint Plot
A scatter plot using the specified columns x and y from the DataFrame data.

A (univariate) histogram along the top of the scatter plot showing distribution of the column x.

A (univariate) histogram along the right of the scatter plot showing distribution of the column y.

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

#### kind of joint plot
The seaborn function sns.jointplot() has a parameter kind to specify how to visualize the joint variation of two continuous random variables

kind='scatter' uses a scatter plot of the data points

kind='reg' uses a regression plot (default order 1)

kind='resid' uses a residual plot

kind='kde' uses a kernel density estimate of the joint distribution

kind='hex' uses a hexbin plot of the joint distribution

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='resid')
```

### Pair plot
Visualize the mutlivariate plot. Grid of joint plot of numeric datatype fields

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='smoker', kind='reg')
```
